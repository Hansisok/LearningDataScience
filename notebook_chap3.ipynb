```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# 定义tanh激活函数
def tanh_activation(x):
    return np.tanh(x)

# 生成数据点
x = np.linspace(-10, 10, 400)
y = tanh_activation(x)

# 绘制tanh激活函数
plt.plot(x, y, label="tanh(x)")
plt.title("tanh")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.grid(True)
plt.legend()
plt.show()
```
